# Exercise 2

Write a function `display_dict(dictionary)` that takes a dictionary argument and displays it in the way shown below.

```python
display_dict({'cat': 'Kot', 'dog': 'Pies', 'bird': 'Ptak'})
```

```
cat=>Kot
dog=>Pies
bird=>Ptak
```

````python
# Write the "display_dict" function here
````

````python
import sys
import io
_real_stdout = sys.stdout

sys.stdout = io.StringIO()
display_dict({'cat': 'Kot', 'dog': 'Pies'})
sys.stdout.seek(0)
if set(sys.stdout.read().strip().split()) == {'cat=>Kot', 'dog=>Pies'}:
    print('Passed!', file=_real_stdout)
else:
    print('display_dict function did not display expected lines', file=_real_stdout)
    print('Function displayed', file=_real_stdout)
    sys.stdout.seek(0)
    print(sys.stdout.read(), file=_real_stdout)
    print('Instead of expected:\ncat=>Kot\ndog=>Pies\n', file=_real_stdout)

sys.stdout = io.StringIO()
display_dict({'bird': 'Ptak', 'mouse': 'Mysz'})
sys.stdout.seek(0)
if set(sys.stdout.read().strip().split()) == {'bird=>Ptak', 'mouse=>Mysz'}:
    print('Passed!', file=_real_stdout)
else:
    print('display_dict function did not display expected lines', file=_real_stdout)
    print('Function displayed', file=_real_stdout)
    sys.stdout.seek(0)
    print(sys.stdout.read(), file=_real_stdout)
    print('Instead of expected:\nbird=>Ptak\nmouse=>Mysz\n', file=_real_stdout)

sys.stdout = _real_stdout
````
