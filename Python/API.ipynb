## Práce s JSON soubory

**JSON (JavaScript Object Notation)** je jednoduchý formát pro výměnu dat, který je snadno čitelný pro lidi a snadno zpracovatelný pro stroje. Používá se pro strukturovanou výměnu dat mezi klientem a serverem.

### Základní struktura JSON:

JSON je tvořen dvěma hlavními strukturami:

**1. Objekty**

**2. Pole**

### Objekty
- Objekty v JSON jsou kolekce párů klíč-hodnota. 
- Každý klíč je textový řetězec (obvykle uváděný v uvozovkách), následuje dvojtečka a hodnota. 
- Hodnoty mohou být různých typů: řetězce, čísla, boolean (true/false), pole nebo jiné objekty.

Příklad objektu:

```python
{
  "jméno": "Jan",
  "věk": 30,
  "město": "Praha"
}
```

### Pole
- **Pole** v JSON jsou seznamy hodnot uzavřených v hranatých závorkách. 
- Hodnoty mohou být jakéhokoli typu a mohou být dokonce i objekty nebo jiná pole.

Příklad pole:

```python
[
  "jablko",
  "banán",
  "třešeň"
]
```

Pole objektů

```python
[
  {
    "jméno": "Jan",
    "věk": 30
  },
  {
    "jméno": "Eva",
    "věk": 25
  }
]
```

### Kombinace objektů a polí
JSON je velmi flexibilní a umožňuje vnořování objektů a polí do sebe.

Příklad kombinace objektu a pole:

```python
{
  "jméno": "Jan",
  "věk": 30,
  "děti": [
    {
      "jméno": "Eva",
      "věk": 5
    },
    {
      "jméno": "Petr",
      "věk": 3
    }
  ]
}
```

### Proč používat JSON?
- **Jednoduchost**: JSON je snadno čitelný a psaný.
- **Jazyková nezávislost**: I když JSON vychází z JavaScriptu, je podporován ve většině moderních programovacích jazyků.
- **Efektivní výměna dat**: JSON je lehký formát, což ho činí ideálním pro výměnu dat mezi klientem a serverem, zvláště v webových aplikacích.
- Pokud si to představíte jako tabulku nebo seznamy, kde máte jasně definované dvojice "klíč-hodnota", JSON bude snadno pochopitelný.

## Serializace a deserializace

`json.dumps`

`Co dělá`: Převádí Python objekt (např. slovník nebo seznam) na JSON řetězec.

`Použití`: Pokud chcete data serializovat, například je uložit do souboru nebo odeslat přes síť.

`json.loads`

`Co dělá: Načítá (deserializuje) JSON řetězec a převádí ho na Python objekt, jako je slovník nebo seznam.
Použití: Pokud přijmete JSON data z API, souboru nebo jiného zdroje a potřebujete s nimi pracovat v Pythonu.

```python
# Deserializace z .json souboru do python struktury (dictionary)
import json
from pprint import pprint
with open('data.json', 'r') as file:
    data = json.load(file)

# Získání aktuální teploty
teplota = data['weatherCondition']['actualTemperature']
# Získání vlhkosti
vlhkost = data['weatherCondition']['humidity']
# Získání města
mesto = data['town']

print(f'Ve městě {mesto} je - Vlhkost: {vlhkost}%, teplota: {teplota}')
```

# API 

API (Application Programming Interface) je sada pravidel, která umožňuje aplikacím vzájemně komunikovat

API definuje metody a datové struktury, které vývojáři mohou použít k integraci aplikací a služeb.

### Principy REST API

HTTP Metody

**GET**: Získání dat ze serveru.

**POST**: Vytvoření nového záznamu na serveru.

**PUT**: Aktualizace existujícího záznamu.

**DELETE**: Smazání záznamu.

**URL Konvence**

**Resourcy**: V RESTful API jsou data reprezentována jako "resourcy" (zdroje), které mají jedinečné URL.

**Příklad**: /users, /users/123

### Status Kódy
**200 OK**: Úspěšné získání dat nebo provedení operace.

**201 Created**: Úspěšné vytvoření nového záznamu.

**400 Bad Request**: Chybný požadavek (např. neplatná data).

**401 Unauthorized**: Neautorizovaný přístup.

**404 Not Found**: Záznam nebyl nalezen.

**500 Internal Server Error**: Chyba na straně serveru.


### Querystring
- Query string je část URL, která obsahuje specifické parametry a hodnoty, které se odesílají serveru jako součást požadavku. 
- Query string se přidává na konec URL a začíná znakem ?. Následují páry klíč-hodnota, které jsou odděleny znakem &.

Příklad URL s query string:

```python
'https://example.com/api/users?age=25&city=Prague'
```

V tomto příkladu:

- `https://example.com/api/users` je základní URL.
- `?` označuje začátek query stringu.
- `age=25` je první pár klíč-hodnota, kde age je klíč a 25 je hodnota.
- `&` odděluje jednotlivé páry klíč-hodnota.
- `city=Prague` je druhý pár klíč-hodnota, kde city je klíč a Prague je hodnota.

### Jak API používá Query String?
##### **Query string** se často používá pro filtrování, třídění nebo stránkování výsledků. Například:

**Filtrování**

API může používat query string k filtrování výsledků podle určitých kritérií:

```python
from requests import get


response = get("https://api.example.com/products?category=electronics&price_lt=1000")
```

Tento požadavek vyžaduje všechny produkty v kategorii "electronics" s cenou nižší než 1000.

### Třídění
Query string může určovat, jakým způsobem budou výsledky tříděny:

```python
`GET https://api.example.com/products?sort=price&order=asc`
```

Tento požadavek třídí produkty podle ceny ve vzestupném pořadí.

### Stránkování
Query string může také definovat stránkování výsledků, tedy rozdělení na více stránek:

```python
`GET https://api.example.com/products?page=2&limit=20`
```

Tento požadavek vyžaduje druhou stránku výsledků, kde každá stránka obsahuje 20 položek.

Příklad: Uživatel zadá měnu, pro kterou chce zíslat směnný kurz z PLZ a my mu vrátíme aktuální kurz

```python
# Funkce pro načtení měny od uživatele
def nacti_menu_od_uzivatele():
    code = input(f'Zadej měnu, pro kterou chceš získat směnný kurz z PLZ: ')
    return code

# Funkce pro zjištění směnného kurzu dané měny
def zjisti_smenny_kurz(mena):
    response = requests.get(f'https://api.nbp.pl/api/exchangerates/rates/a/{mena}')
    data = response.json()
    return data['rates'][0]['mid']

# Funkce, která od uživatele získá částku k převodu
def nacti_castku_od_uzivatele():
    castka = int(input('Zadej částku, kterou chceš převést na PLN: '))
    return castka

# Funkce, která mi dle směnného kurzu převede zadanou částku na PLN
def vypocitej_smenenou_castku(kurz, castka):
    return kurz * castka
```

```python
mena = nacti_menu_od_uzivatele() # Získává od uživatele měnu k převodu
kurz = zjisti_smenny_kurz(mena) # Volá API
print(f'Směnný kurz z {mena} na PLN je: {kurz}')

castka_k_prevodu = nacti_castku_od_uzivatele() # Načítáme částku k převodu od uživatele
castka_smenena = vypocitej_smenenou_castku(kurz, castka_k_prevodu) # Převádíme danou částku na PLN

print(f'{castka_k_prevodu} {mena} je dle aktuálního směnného kurz, jenž činí {kurz}, přesně {castka_smenena} PLN')
```
